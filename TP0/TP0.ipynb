```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

### Création et analyse des données

```python
X = np.array([[6], [8], [10], [14], [18]])
y = [7, 9, 13, 17.5, 18]
```

```python
plt.scatter(X,y)
plt.title("Pizzas prices plotted against sizes")
plt.xlabel("Sizes in cms")
plt.ylabel("Prices in Euros")
plt.grid()
plt.show()
```

### Regression linéaire

```python
model = LinearRegression()
model.fit(X,y)
pred = model.predict(X)
```

```python
plt.scatter(X,y)
plt.plot(X,pred,c="red")
plt.title("Pizzas prices plotted against sizes")
plt.xlabel("Sizes in cms")
plt.ylabel("Prices in Euros")
plt.grid()
plt.show()
```

### Evaluation du modèle

```python
error =   ((y - pred)**2).sum()
print("Residual sum of squares : " , error)
```

### Etude théorique de notre problème de pizza.

```python
# Variance
variance = np.var(X,ddof=1)

#Covariance
covariance = np.cov(X.transpose(), y)[0][1]

# Alpha
alpha = covariance/variance

# beta
beta = np.mean(y) - alpha*np.mean(X) 

alpha,beta
```

```python
# plot
pred2 = alpha*X + beta
plt.scatter(X,y)
plt.plot(X,pred2,c="red")
plt.title("Pizzas prices plotted against sizes")
plt.xlabel("Sizes in cms")
plt.ylabel("Prices in Euros")
plt.grid()
plt.show()
```

```python
X_test = np.array([[8], [9], [11], [16], [12]])
y_test = [11, 8.5, 15, 18, 11]
model.score(X_test,y_test)
```
